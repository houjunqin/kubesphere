##<font color='red'>Part-1</font><br>
Create three lists: (1) core_list, (2) elective_list, and (3) capstone_list. The list elements are the course numbers line “DATA 5050”. Do not include the course titles.
*  	Print the contents of each list
*  Print the number of elements in each list using len() method
*  Print the minimum and maximum elements of each list using min() and max() methods.
*  	Perform an appropriate slicing to print only elective courses from “DATA 5900” to COMP 5800”.
*  Sort elective_list and print the sorted list
*  	Check if “DATA 5350” is in core_list using in membership operator. Print True if it is in core_list, otherwise print False. DO not use “if/elif/else” conditional structures.
*  Replace “DATA 5100” in core_list with “DATA 7100” using appropriate index of core_list. Do you get any error message? Explain.
*  	Concatenate core_list, elective_list, and capstone_list to create a single list called DS_list. You can use append() or extend() methods.


```python
core_list = ["DATA 5050","DATA 5100","COMP 5850","COMP 6200"]
elective_list = ["DATA 5200","DATA 5300","DATA 5400","DATA 5500","DATA 5900","DATA 6100","DATA 6150","DATA 5350","COMP 5400","COMP 5800","COMP 6400","COMP 6800"]
capstone_list = ["DATA 6200"]
# Print the contents of each list
print("Core Courses:",core_list)
print("Elective Courses:",elective_list)
print("Capstone Courses:",capstone_list)

# Print the number of elements in each list using len() method
print("The number of elements : core_list=%s, elective_list=%s, capstone_list=%s." %(len(core_list),len(elective_list),len(capstone_list)))

# Print the minimum and maximum elements of each list using min() and max() methods
print(f"Maximum element in list: core_list - {max(core_list)}, elective_list - {max(elective_list)}, capstone_list - {max(capstone_list)}")
print("Minimum element in list: core_list - {}, elective_list - {}, capstone_list - {}".format(min(core_list),min(elective_list),min(capstone_list)))

# Perform an appropriate slicing to print only elective courses from “DATA 5900” to COMP 5800”.
print("Perform slicing from \"DATA 5900\" to \"COMP 5800\":",elective_list[4:-2])

# Sort elective_list and print the sorted list
print("Sorted elective_list:", sorted(elective_list,reverse=False))

# Check if “DATA 5350” is in core_list using in membership operator. Print True if it is in core_list, otherwise print False. 
# DO not use “if/elif/else” conditional structures
print("Check if \"DATA 5350\" is in core_list: ","DATA 5350" in core_list)

# Replace “DATA 5100” in core_list with “DATA 7100” using appropriate index of core_list. Do you get any error message? Explain.
# Explain: I don't get any error because the list is mutable.
core_list[1] = "DATA 7100"
print("The updated core_list is:",core_list)

# Concatenate core_list, elective_list, and capstone_list to create a single list called DS_list. You can use append() or extend() methods.
DS_list = []
DS_list.extend(core_list)
DS_list.extend(elective_list)
DS_list.extend(capstone_list)
print("The concatonation of core_list, elective_list, and capstone_list is:",DS_list)
```

##<font color='red'>Part-2</font><br>
Create three tuples: (1) core_tuple, (2) elective_tuple, and (3) capstone_tuple. The tuple elements are the course numbers like “DATA 5050”. Do not include the course titles.
*  	Print the contents of each tuple
*  	Print the number of elements in each list using len() method
*  	Print the minimum and maximum elements of each list using min() and max() methods.
*  	Perform an appropriate slicing to print only elective courses from “DATA 5900” to COMP 5800”.
*  	Sort elective_tuples and print the sorted tuple
*  	Check if “DATA 5350” is in core_tuple using in membership operator. Print True if it is in core_list, otherwise print False. DO not use “if/elif/else” conditional structures.
*  	Replace “DATA 5100” in core_tuple with “DATA 7100” using appropriate index of core_tuple. Do you get any error message? Explain.
*  	Can you concatenate core_tuple, elective_tuple, and capstone_tuple to create a single tuple called DS_tuple? Explain

```python
core_tuple = ("DATA 5050","DATA 5100","COMP 5850","COMP 6200")
elective_tuple = ("DATA 5200","DATA 5300","DATA 5400","DATA 5500","DATA 5900","DATA 6100","DATA 6150","DATA 5350","COMP 5400","COMP 5800","COMP 6400","COMP 6800")
capstone_tuple = ("DATA 6200")

# Print the contents of each tuple
print("Core Courses:",core_tuple)
print("Elective Courses:",elective_tuple)
print("Capstone Courses:",capstone_tuple)

# Print the number of elements in each list using len() method
print("The number of elements : core_tuple=%s, elective_tuple=%s, capstone_tuple=%s." %(len(core_tuple),len(elective_tuple),len(capstone_tuple)))

# Print the minimum and maximum elements of each list using min() and max() methods.
print(f"Maximum element in tuple: core_tuple - {max(core_tuple)}, elective_tuple - {max(elective_tuple)}, capstone_tuple - {max(capstone_tuple)}")
print("Minimum element in tuple: core_tuple - {}, elective_tuple - {}, capstone_tuple - {}".format(min(core_tuple),min(elective_tuple),min(capstone_tuple)))

# Perform an appropriate slicing to print only elective courses from “DATA 5900” to COMP 5800”
print("Perform slicing from \"DATA 5900\" to \"COMP 5800\":",elective_tuple[4:-2])

# Sort elective_tuples and print the sorted tuple
print("Sorted elective_tuple:", sorted(elective_tuple,reverse=False))

# Check if “DATA 5350” is in core_tuple using in membership operator. Print True if it is in core_list, otherwise print False. 
# DO not use “if/elif/else” conditional structures
print("Check if \"DATA 5350\" is in core_tuple: ","DATA 5350" in core_tuple)

# Can you concatenate core_tuple, elective_tuple, and capstone_tuple to create a single tuple called DS_tuple? Explain
# Explain: I can't create the tuple directly because tuples are immutable because you cannot append, delete, or edit tuples at all.
# But I can convert every tuple into list, then I use append or extend method to concatenate the list, final I convert the list back to tuple.
templist = []
templist.extend(list(core_tuple))
templist.extend(list(elective_tuple))
templist.extend(list(capstone_tuple))
DS_tuple = tuple(templist)
print("The concatonation of core_tuple, elective_tuple, and capstone_tuple is:\n",DS_tuple)

# Replace “DATA 5100” in core_tuple with “DATA 7100” using appropriate index of core_tuple. Do you get any error message? Explain.
# Explain: Yes, I get the error message because the tuple is immutable.
core_tuple[1] = "DATA 7100"
print("The updated core_tuple is:",core_tuple)
```

##<font color='red'>Part-3</font><br>
Create three dictionaries: (1) core_dictionary, (2) elective_dictionary, and (3) capstone_dictionary. The dictionary keys are the course numbers like “DATA 5050” and values are the course titles like “Programming for Data Science”. 
* 	Print the contents of each dictionary
* 	Print the keys of each dictionary
* 	Print the value that corresponds to the key “COMP 6400” in elective_dictionary.
* 	Add a new element to core_dictionary with the key “DATA 5950” and the value “Unknown Course”. Print core_dictionary after adding the new element.
* 	Add a repeated element to the core_dictionary. For example, add a new element with the key “DATA 5100” and the value “Programming for Data Science”. What do you observe? Explain.
* 	Pop a random element from elective_dictionary using pop() method and print it.

```python
core_dictionary = {"DATA 5050":"Mathematics for Data Science","DATA 5100": "Programming for Data Science",
                   "COMP 5850": "Data Visualization","COMP 6200": "Machine Learning"}
elective_dictionary = {"DATA 5200": "Statistical Learning","DATA 5300": "Data Mining",
                       "DATA 5400": "Algorithms for Data Science","DATA 5500": "Business Data Analytics",
                       "DATA 5900": "Special Topics","DATA 6100": "Natural Language Processing",
                       "DATA 6150": "Applied Deep Learning","DATA 5350": "Applied Statistics for Data Science",
                       "COMP 5400": "Hybrid and Relational Databases","COMP 5800": "Introduction to Bioinformatics",
                       "COMP 5850": "Data Visualization","COMP 6400": "Distributed Algorithm Design and Data Analysis",
                       "COMP 6800": "Introduction to Computer Vision"}
capstone_dictionary = {"DATA 6200": "Data Science Capstone"}

# Print the contents of each dictionary
print("Core Courses:",core_dictionary)
print("Elective Courses:",elective_dictionary)
print("Capstone Courses:",capstone_dictionary)

# Print the keys of each dictionary
print(core_dictionary.keys())
print(elective_dictionary.keys())
print(capstone_dictionary.keys())

# Print the value that corresponds to the key “COMP 6400” in elective_dictionary.
print(elective_dictionary['COMP 6400'])

# Add a new element to core_dictionary with the key “DATA 5950” and the value “Unknown Course”. Print core_dictionary after adding the new element.
core_dictionary["DATA 5950"]="Unknown Course"
print(core_dictionary)

# Add a repeated element to the core_dictionary. For example, add a new element with the key “DATA 5100” and the value “Programming for Data Science”.
# What do you observe? Explain.
# The key is unique in dictionary so you can't add the same key more than once. The new value will overwrite the old value if the key exists. 
core_dictionary["DATA 5100"]="Programming for Data Science"
print(core_dictionary)

# Pop a random element from elective_dictionary using pop() method and print it.
print(elective_dictionary.pop("COMP 5850"))
```
